```python
import numpy as np
import matplotlib.pyplot as plt
import math
from matplotlib.pyplot import rcParams
rcParams.update({'font.size':20}) #optional to change default font size

# ---------- define constants

q = 1e-6    # Coulombs
m = 0.1     # kilograms
k = 8.99e9  # Newton meter squared per Coulomb
r = 1       # meters

# ---------- set up TLF function

tin = 0           # initial time
tfin = 1000       # final time
dt = 1
half_dt = dt/2

TLF = np.arange(tin, tfin, dt)

def force(x):
    return ((-q)*k*q*x)/((x**2+r**2)**(3/2))

def iterate(x0):  
    x = x0
    v = 0
    b = np.zeros(len(TLF))
    for i in range(len(TLF)):
        v = v + (force(x)/m) * half_dt
        x = x + v*dt
        v = v + (force(x)/m) * half_dt
        b[i] = x
    return b

# ---------- plot trajectories
%matplotlib inline

# x0 = 1 m
b1 = iterate(1)
plt.plot(TLF, b1)
plt.title('Position as a function of time for x0 = 1m')
plt.xlabel('time, t (s)')
plt.ylabel('position, x (m)')
plt.show()

# x0 = 10 m
b10 = iterate(10)
plt.plot(TLF, b10)
plt.title('Position as a function of time for x0 = 10m')
plt.xlabel('time t (s)')
plt.ylabel('position x (m)')
plt.show()
```

difference between two graphs: x0 = 1m has a shorter period and smaller amplitude than x0 = 10m
